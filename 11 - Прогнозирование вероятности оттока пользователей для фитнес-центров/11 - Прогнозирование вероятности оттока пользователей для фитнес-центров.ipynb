# Проект - модель предсказания оттока клиентов фитнес-центра

## Описание проекта

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.
***
Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент больше не с вами? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.
***
Индикаторы оттока зависят от специфики отрасли. Когда пользователь редко, но стабильно закупается в интернет-магазине — не похоже, что он «отвалился». А вот если две недели не заходит на канал с ежедневно обновляемым контентом, дела плохи: подписчик заскучал и, кажется, оставил вас.
***
Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.
Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Ваша задача — провести анализ и подготовить план действий по удержанию клиентов.
***
А именно:
- научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать основные признаки, наиболее сильно влияющие на отток;
- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
    1) выделить целевые группы клиентов;
    2) предложить меры по снижению оттока;
    3) определить другие особенности взаимодействия с клиентами.

«Культурист-датасаентист» предоставил сведения в csv-файлах. Заказчик подготовил данные, которые содержат данные на месяц до оттока и факт оттока на определённый месяц. Набор данных включает следующие поля:
***
Данные клиента за предыдущий до проверки факта оттока месяц:
- 'gender' — пол;
- 'Near_Location' — проживание или работа в районе, где находится фитнес-центр;
- 'Partner' — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент в таком случае фитнес-центр хранит информацию о работодателе клиента);
- Promo_friends — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
- 'Phone' — наличие контактного телефона;
- 'Age' — возраст;
- 'Lifetime' — время с момента первого обращения в фитнес-центр (в месяцах).
***
Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
- 'Contract_period' — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
- 'Month_to_end_contract' — срок до окончания текущего действующего абонемента (в месяцах);
- 'Group_visits' — факт посещения групповых занятий;
- 'Avg_class_frequency_total' — средняя частота посещений в неделю за все время с начала действия абонемента;
- 'Avg_class_frequency_current_month' — средняя частота посещений в неделю за предыдущий месяц;
- 'Avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.
***
- 'Churn' — факт оттока в текущем месяце.

# Построение моделей

## Классификация

### Random forest

```python
model_data = data.drop(['month_to_end_contract', 'avg_class_frequency_current_month'], axis=1)
X = model_data.drop('churn', axis=1) #датафрейм с фичами
y = model_data['churn'] # таргет
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf_model = RandomForestClassifier(n_jobs=-1, random_state=42)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)
```

```python
scores = ['accuracy', 'precision', 'recall']
print("Значения метрик на кросс-валидации для базовой модели Random Forest")
for score in scores:
    print(f'Cross-validated {score} = ',cross_val_score(rf_model, X, y, cv=5, scoring=score).mean())
sns.set(rc={'figure.figsize': (8, 4)})
ConfusionMatrixDisplay(confusion_matrix(y_test, rf_preds)).plot()
plt.grid();
```

```python
print(classification_report(y_test, rf_preds))
```

```python
importances = pd.DataFrame({'Признаки':X.columns})
importances['Важность'] = pd.Series(rf_model.feature_importances_)
importances
```

#### Настройка гиперпараметров случайного леса

```python
grid = {'n_estimators':range(10, 70, 10),
        'criterion':['entropy', 'gini'],
       'max_depth':range(5, 30, 5),
       'min_samples_leaf':range(2,6),
       'min_samples_split':range(2,6)} #словарь со значениями гиперпараметров

rf_gs = RandomForestClassifier(n_jobs=-1,random_state=42)
rf_model_cv = RandomizedSearchCV(param_distributions=grid,
                                estimator=rf_gs,
                                n_iter=50,
                                cv=5)
rf_model_cv.fit(X_train, y_train)
```

```python
best_rf_model = rf_model_cv.best_estimator_

best_rf_model.fit(X_train, y_train)
best_rf_model_preds = best_rf_model.predict(X_test)

# {'n_estimators': 30,
#  'min_samples_split': 4,
#  'min_samples_leaf': 3,
#  'max_depth': 15,
#  'criterion': 'gini'}
```

```python
print("Значения метрик на кросс-валидации для модели Random Forest после тюнинга гиперпараметров")
for score in scores:
    print(f'Cross-validated {score} = ', cross_val_score(best_rf_model, X, y, cv=5, scoring=score).mean())

ConfusionMatrixDisplay(confusion_matrix(y_test, best_rf_model_preds)).plot()
plt.grid();
```

```python
print(classification_report(y_test, best_rf_model_preds))
```

```python
importances = pd.DataFrame({'Признаки':X.columns})
importances['Важность'] = pd.Series(best_rf_model.feature_importances_)
importances
```

### Logistic Regression

```python
log_reg_model = LogisticRegression(solver='liblinear', random_state=42)
log_reg_model.fit(X_test, y_test)
log_reg_model_preds = log_reg_model.predict(X_test)
```

```python
scores = ['accuracy', 'precision', 'recall']
print("Значения метрик на кросс-валидации для baseline логистической регрессии")
for score in scores:
    print(f'Cross-validated {score} = ',cross_val_score(log_reg_model, X, y, cv=5, scoring=score).mean())

ConfusionMatrixDisplay(confusion_matrix(y_test, log_reg_model_preds)).plot()
plt.grid();
```

```python
print(classification_report(y_test, log_reg_model_preds))
```

```python
for i in log_reg_model.coef_:
    a=i
importances = pd.DataFrame({'Признаки':X.columns})
importances['Важность'] = pd.Series(a)
importances
```

#### Настройка гиперпараметров логистической регрессии

```python
grid = {'solver':['liblinear'],#sag, saga
        'penalty':['l1', 'l2'],
        'C':range(1,5),
       'max_iter':range(100, 1000)}
       

clf = LogisticRegression(random_state=42)
logreg_model_cv = RandomizedSearchCV(param_distributions=grid,
                                estimator=clf,
                                n_iter=50,
                                cv=5,
                                error_score=0)
logreg_model_cv.fit(X_train, y_train)
logreg_model_cv_preds = logreg_model_cv.predict(X_train)
```

```python
best_logreg = logreg_model_cv.best_estimator_
best_logreg_preds = best_logreg.predict(X_test)
scores = ['accuracy', 'precision', 'recall']
print("Значения метрик на кросс-валидации для логистической регрессии после тюнинга гиперпараметров")
for score in scores:
    print(f'Cross-validated {score} = ',cross_val_score(best_logreg, X, y, cv=5, scoring=score).mean())
ConfusionMatrixDisplay(confusion_matrix(y_test, best_logreg_preds)).plot()
plt.grid();
```

```python
print(classification_report(y_test, best_logreg_preds))
```

```python
for i in best_logreg.coef_:
    a=i
importances = pd.DataFrame({'Признаки':X.columns})
importances['Важность'] = pd.Series(a)
importances
```

***
Для наших целей наиболее подходящей является последняя модель логистической регрессии - она наиболее точно предсказывает именно  факт ухода клиентов (recall = 0.79), но с предсказанием статуса действующих клиентов справляется несколько хуже остальных моделей

## Кластеризация

### Агломеративная иерархическая кластеризация

```python
X = data.drop('churn', axis=1)
```

```python
sc = StandardScaler() 
X_sc = sc.fit_transform(X) # стандартизируем фичи 
```

```python
linked = linkage(X_sc, method='ward')
```

```python
plt.figure(figsize=(15, 7))  
dendrogram(linked, orientation='top')
plt.title('Hierarchial clustering for GYM') # построим дендрограмму
plt.show() 
```

***
При агломеративной иерархической кластеризации выделяется 4 кластера
***

### KMeans

```python
km = KMeans(n_clusters=5, random_state=42)
labels = km.fit_predict(X_sc)
model_data['clusters'] = labels
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_sc, labels) # оценим качество кластеризации
```

```python
plt.figure(figsize=(20,5)) # средние значения признаков в кластерах
sns.heatmap(model_data.groupby('clusters').mean(), annot=True, fmt='.3g', cbar=False);
```

***
В 1 и 2 кластерах наименьшее число ушедших клиентов - меньше 1 % (из 889 и 865, соответственно)
***
2 кластер самый лояльный - они тратят на доп.услуги больше всех, живут в основном рядом с фитнес-центром, работают в компаниях-партнерах, почти половина из них пришла по промо-акции (видимо, поэтому их средняя продолжительность действия абонемента - почти целый год)
***
1 кластер отличается от 2 гораздо более низкой долей сотрудников партнеров фитнес-центра, а по промо-акции из них пришло лишь 5 %. Абонементы они покупают на куда меньший срок, 2.5 месяца. При этом занятия они посещают несколько чаще.
***
Наибольшее число ушедших клиентов - в 3 кластере, около 57%. Это самый многочисленный кластер, 1227 человек. Среди них меньшее число живут в районе фитнес-центра. Их лайфтайм меньше, чем в остальных кластерах, и они покупают абонементы на меньший срок, реже посещают занятия, меньше тратят на доп. услуги.
***
0 и 4 кластеры очень похожи друг на друга, единственные серьезные отличия - в том, как именно они становятся клиентами (сотрудники партнеров, промо-акция) и длительности абонементов. А еще клиенты из 4 кластера не оставляют номера телефона (вероятно, это стало одной из причин выделения именно этих клиентов в отдельный кластер)
